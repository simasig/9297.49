```python
#Scatter Plot - Carat vs Price:This code generates a scatter plot depicting the relationship between the carat weight of diamonds and their prices
sb.scatterplot(data=df, x='carat', y='price')
plt.title('Scatter Plot - Carat vs Price')
plt.show()
```

```python
#Box Plot - Price distribution by Cut:This code creates a box plot to visualize the distribution of diamond prices based on their cut quality.
sb.boxplot(data=df, x='cut', y='price')
plt.title('Box Plot - Price distribution by Cut')
plt.show()
```

```python
#Histogram - Price Distribution:This code generates a histogram illustrating the distribution of diamond prices, including a kernel density estimate.
sb.histplot(data=df, x='price', bins=30, kde=True)
plt.title('Histogram - Price Distribution')
plt.show()
```

```python
#Pair Plot - Comparison of Diamond Dimensions and Price:This code produces a pair plot for a subset of diamond dimensions, providing a visual comparison.
sb.pairplot(df[['carat', 'depth', 'table', 'price']])
plt.suptitle('Pair Plot - Comparison of Diamond Dimensions and Price', y=1.02)
plt.show()
```

```python
correlation_matrix = df[['carat', 'depth', 'table', 'price']].corr()
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap - Correlation Matrix')
plt.show()
```
